```python
plt.subplot(2, 2, 1)
plt.hist(dataset['alcohol'], bins=20)
plt.xlabel('Alcohol')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.subplot(2, 2, 3)
correlation_matrix = dataset.corr()
plt.imshow(correlation_matrix, cmap='coolwarm', interpolation='nearest')
plt.colorbar()
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.subplot(2, 2, 4)
plt.scatter(dataset['volatile acidity'], dataset['citric acid'])
plt.xlabel('Volatile Acidity')
plt.ylabel('Citric Acid')
plt.show()
```

## Report on Graphs Obtained

---

The red wine dataset encompasses a range of attributes related to red wine, including *alcohol content, pH level, quality, volatile acidity, and citric acid*. Exploring the dataset yielded valuable insights into these attributes.

### **Graphs**

**_Alcohol Content Histogram_**:
   - The histogram highlighted that the red wine's alcohol content mostly falls within the 9% to 15% range.
   - The distribution showed a somewhat right-skewed pattern, indicating that wines with higher alcohol content are less common.


**_Correlation Heatmap_**:
- The heatmap depicted correlations between different features in the dataset.
- Some notable correlations were observed, such as a negative correlation between volatile acidity and citric acid.

**_Scatter Plot of Volatile Acidity vs. Citric Acid_**:
- The scatter plot showed a moderate negative correlation between volatile acidity and citric acid.
- Wines with higher citric acid content tend to have lower volatile acidity and vice versa.

## Insights:
1._Alcohol Content_: Red wines predominantly exhibit an alcohol content ranging from 9% to 15%, with fewer instances of higher alcohol content wines.

2._pH and Quality_: Wines with higher quality ratings appear to have slightly lower pH levels, suggesting a potential correlation between quality and acidity.

3._Correlations_: Some attributes displayed notable correlations, offering insights into potential relationships between features.

4._Volatile Acidity and Citric Acid_: A moderate negative correlation between volatile acidity and citric acid suggests a potential relationship between these attributes.
